```python
import matplotlib.pyplot as plt
```

```python
# Defining Function
def f(x1,x2):
    return (x1-2)**2 + (x2-2)**2

# Defining derivative of function
def g(x):
    return 2*(2*x-4)
# Implementing Newton Raphson Method
CN1=[]
CN2=[]
I=[]
def newtonRaphson(x1,x2,e,N):
    print('\n\n** NEWTON RAPHSON METHOD IMPLEMENTATION **')
    step = 0
    flag = 1
    condition = True
    while condition:
        if g(x1) == 0.0:
            print('Divide by zero error!')
            break
        if g(x2) == 0.0:
            print('Divide by zero error!')
            break
        CN1.append(x1)
        CN2.append(x2)
        I.append(step)
        x11 = x1 - f(x1,x2)/g(x1)
        x22 = x2 - f(x1,x2)/g(x2)
        print('Iteration-%d, x1 = %0.6f,x2 = %0.6f, and f(x1,x2) = %0.6f' % (step, x11,x22, f(x11,x22)))

        x1 = x11
        x2 = x22
        step = step + 1


        if step > N:
            flag = 0
            break

        condition = abs(f(x1,x2)) > e

    if flag==1:
        print('\nRequired root is: %0.8f' % x1,x2)
    else:
        print('\nNot Convergent.')



# Input Section
# Converting x0 and e to float
# Converting N to integer
x1 = float(input('Enter Guess: '))
x2 = float(input('Enter Guess: '))
e = float(input('Tolerable Error: '))
N = int(input('Maximum Step: '))

# Starting Newton Raphson Method
newtonRaphson(x1,x2,e,N)

print("convergence criteria=",CN1,CN2)
print("iterations",I)
plt.plot(I,CN1,CN2,label = r'Newton Raphson Method')
plt.xlabel('iterations')
plt.ylabel('convergence criteria')
plt.legend()
```

```python
Cg1=[]
Cg2=[]
Ig=[]
def gradient_descent(x1,x2, learn_rate, n_iter):

    for N in range(n_iter):
        Cg1.append(x1)
        Cg2.append(x2)
        Ig.append(N)
        x11 = x1-(learn_rate * g(x1))
        x22 = x2-(learn_rate * g(x2))
        x1 = x11
        x2 = x22


# Input Section
def g(x):
    return 2*(2*x-4)

start = input('Enter Guess: ')
start2 = input('Enter Guess: ')
learn_rate = input('learn_rate: ')
n_iter = input('Maximum Step: ')

# Converting x0 and e to float
x1 = float(start)
x2 = float(start2)
learn_rate = float(learn_rate)
# Converting N to integer
n_iter = int(n_iter)

gradient_descent(x1,x2,learn_rate,n_iter)
print("convergence criteria=",Cg1,Cg2)
print("iterations",Ig)
plt.plot(Ig, Cg1,Cg2,label = r'Gradient Descent Method')
plt.xlabel('iterations')
plt.ylabel('convergence criteria')
plt.legend()
```

```python
# print and plot the convergence criteria versus the number of iterations for the Gradient Descent Method and Newton Raphson Method
print("Cg X1=",Cg1)
print("CN X1=",CN1)
print("Cg X2=",Cg2)
print("CN X2=",CN2)

plt.plot(I,CN1,CN2,label = r'Newton Raphson Method')
plt.plot(Ig, Cg1,Cg2,ls = '--',label = r'Gradient Descent Method')
plt.xlabel('iterations')
plt.ylabel('convergence criteria')
plt.legend()
plt.title('point 2')
```
